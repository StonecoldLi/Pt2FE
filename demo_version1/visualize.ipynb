```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#平面系数
planes = [(-8.45630394e-04, -9.50798637e-01,  3.09808388e-01,  3.10613170e+00), (-3.22005644e-03, -1.11233907e-03,  9.99994197e-01, -1.80736717e+00), (1.03347937e-03, -9.47036720e-01, -3.21123626e-01, -1.87959690e+00)]
```

```python
# 创建数据点
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# 创建3D图形
fig = plt.figure()
plt3d = fig.add_subplot(111, projection='3d')

# 绘制多个平面
for plane in planes:
    A, B, C, D = plane
    z = (-A * x - B * y - D) / C
    plt3d.plot_surface(x, y, z, alpha=0.5)

# 设置坐标轴标签
plt3d.set_xlabel('X')
plt3d.set_ylabel('Y')
plt3d.set_zlabel('Z')

# 显示图形
plt.show()
```
